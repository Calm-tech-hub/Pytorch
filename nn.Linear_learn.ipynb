```python
import torch.nn as nn
import torchvision
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
import torch
```

```python
test_dataset = torchvision.datasets.CIFAR10(root = "./dataset",train = False,transform = torchvision.transforms.ToTensor(),download = True)

dataloader = DataLoader(test_dataset,batch_size = 64,drop_last = True)
```

```python
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui,self).__init__()
        self.Linear = nn.Linear(196608,10)

    def forward(self,input):
        output = self.Linear(input)
        return output

tudui = Tudui()
for data in dataloader:
    imgs,targets = data
    print(imgs.shape)
    imgs = torch.flatten(imgs)
    print(imgs.shape)
    output = tudui(imgs)
    print(output.shape)
    
```
