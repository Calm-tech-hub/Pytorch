```python
#最大池化：下采样 MaxPool2d
'''
Parameters:
1.kernel_size:池化核size
2.stride:注意:默认值是kernel_size的大小!!!!!!!!!!!!!!!!!!!!
3.padding:填充
4.dilation:空洞卷积
5.ceil_mode:True:如果池化核里面没有填满，依旧保留，False:不保留

'''
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.nn import MaxPool2d
import torchvision
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
```

```python
input = torch.tensor([[1,2,0,3,1],
                     [0,1,2,3,1],
                     [1,2,1,0,0],
                     [5,2,3,1,1],
                     [2,1,0,1,1]],dtype = float)

input = torch.reshape(input,(1,1,5,5))

```

```python
test_dataset = torchvision.datasets.CIFAR10(root = "./dataset",train = False,transform = torchvision.transforms.ToTensor(),download = True)

dataloader = DataLoader(test_dataset,batch_size = 64)
```

```python
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui,self).__init__()
        self.max_pool = MaxPool2d(kernel_size = 2,ceil_mode = True)

    def forward(self,input):
        output = self.max_pool(input)
        return output
    
tudui = Tudui()
output = tudui(input)
print(output)
#!!!!!很重要：数据类型导致报错！最好还是用float,就是在定义input的时候dtype = float!!!!!!!
```

```python
count = 0
writer = SummaryWriter("pool_pic")
for data in dataloader:
    imgs,targets = data
    print(imgs.shape)
    output = tudui(imgs)
    print(output.shape)
    writer.add_images("pool_pic",output,count)
    count+=1

writer.close()

print(count)

```
