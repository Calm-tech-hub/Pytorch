```python
import torch.nn as nn
import torchvision
import torch
from torch.utils.tensorboard import SummaryWriter
```

```python
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui,self).__init__()
        self.conv1 = nn.Conv2d(3,32,kernel_size = 5,padding = 2)
        self.max_pool1 = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(32,32,5,padding = 2)
        self.max_pool2 = nn.MaxPool2d(2)
        self.conv3 = nn.Conv2d(32,64,5,padding = 2)
        self.max_pool3 = nn.MaxPool2d(2)
        self.flatten = nn.Flatten()
        self.Linear1 = nn.Linear(64*16,64)
        self.Linear2 = nn.Linear(64,10)

    def forward(self,x):
        x = self.conv1(x)
        x = self.max_pool1(x)
        x = self.conv2(x)
        x = self.max_pool2(x)
        x = self.conv3(x)
        x = self.max_pool3(x)
        x = self.flatten(x)
        x = self.Linear1(x)
        x = self.Linear2(x)
        return x
    
tudui = Tudui()
print(tudui)

```

```python
input = torch.randn((64,3,32,32))
output = tudui(input)
print(output)
print(output.shape)
print(output.ndim)
```

```python
#也可用：Sequetial简化代码
class Calm(nn.Module):
    def __init__(self):
        super(Calm,self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(3,32,5,padding = 2),
            nn.MaxPool2d(2),
            nn.Conv2d(32,32,5,padding = 2),
            nn.MaxPool2d(2),
            nn.Conv2d(32,64,5,padding = 2),
            nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Linear(64*16,64),
            nn.Linear(64,10)

        )

    def forward(self,input):
        output = self.model(input)
        return output

calm = Calm()    
input = torch.randn((64,3,32,32))
output = calm(input)
print(output)
print(output.shape)
print(output.ndim)
```

```python
writer = SummaryWriter("Sequential_net")

writer.add_graph(calm,input)

writer.close()
```
