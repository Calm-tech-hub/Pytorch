```python
#准备数据集
import torchvision
import torch
from torch.utils.data import DataLoader
import torch.nn as nn
```

```python
train_dataset = torchvision.datasets.CIFAR10(root = "./dataset",train = True,transform = torchvision.transforms.ToTensor(),download = True )

test_dataset = torchvision.datasets.CIFAR10(root = "./dataset",train = False,transform = torchvision.transforms.ToTensor(),download = True )

train_dataloader = DataLoader(train_dataset,batch_size = 64,shuffle = True)

test_dataloader = DataLoader(test_dataset,batch_size = 64,shuffle = True)
```

```python
train_data_size = len(train_dataset)
print(f"训练集的长度是{train_data_size}")
test_data_size = len(test_dataset)
print(f"测试集的长度是{test_data_size}")
```

```python
from Calm_net import Calm

calm = Calm()
print(calm)

```

```python
#损失函数
loss_fn = nn.CrossEntropyLoss()

#优化器
learning_rate = 0.01
optimizer = torch.optim.SGD(calm.parameters(),lr = learning_rate)
```

```python
epoch_num = 10
total_train_num = 0
total_test_num = 0

for epoch in range(epoch_num):

    #开始训练
    print(f"---------------第{epoch + 1}轮训练开始--------------")
    calm.train()
    for data in train_dataloader:
        imgs,targets = data
        outputs = calm(imgs)
        train_loss = loss_fn(outputs,targets)
        optimizer.zero_grad()
        train_loss.backward()
        optimizer.step()
        total_train_num += 1
        if total_train_num%100 == 0:
            print(f"训练次数:{total_train_num},训练损失：{train_loss}")

    #开始测试
    print(f"---------------第{epoch + 1}轮测试开始--------------")
    calm.eval()
    total_loss = 0
    total_accuracy = 0
    for data in test_dataloader:
        imgs,targets = data
        outputs = calm(imgs)
        accuracy = outputs.argmax(1)
        accuracy = (accuracy == targets).sum()
        total_accuracy += accuracy
        test_loss = loss_fn(outputs,targets).item()
        total_loss+=test_loss

    print(f"整体测试集上的loss:{total_loss}")
    print(f"整体测试集上的accuracy:{total_accuracy/len(test_dataset)}")

    





    
        


    
```
