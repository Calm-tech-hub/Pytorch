```python
import torch.nn as nn
import torchvision
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
```

```python
'''
1.relu
2.sigmoid
3.tanh

'''
```

```python
test_dataset = torchvision.datasets.CIFAR10(root = "./dataset",train = False,transform = torchvision.transforms.ToTensor(),download = True)

dataloader = DataLoader(test_dataset,batch_size = 64,drop_last = True)
```

```python
class Tudui(nn.Module):
    def __init__(self):
        super(Tudui,self).__init__()
        self.relu = F.relu
        self.sigmoid = F.sigmoid

    def forward(self,input):
        input = self.relu(input)
        output = self.sigmoid(input)
        return output
    
writer = SummaryWriter("relu_pic")
tudui = Tudui()
count = 0
for data in dataloader:
    imgs,targets = data
    print(imgs.shape)
    writer.add_images("input",imgs,count)
    output = tudui(imgs)
    writer.add_images("relu_pic",output,count)
    count += 1

print(count)
    
```
